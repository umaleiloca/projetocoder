```python
#pip install requests necessário
import pandas as pd 
import requests
import json
```

```python
# Acessar API1 IBGE UF 
url = "https://servicodados.ibge.gov.br/api/v1/localidades/estados"
response = requests.get(url)
```

```python
# Verificar acesso a API1 IBGE UF 
if response.status_code == 200:
    data_json = response.json()
else:
    print("Erro ao acessar a API")
```

```python
# Ver informações API1 IBGE UF
data_json
```

```python
# Extrair dados API1 IBGE UF em tabela - verificar dados
uf_full = pd.DataFrame(data_json)
uf_full.head(3)
```
