```python
#pip install requests necessário
import pandas as pd 
import requests
import json
```

```python
# API1 FIPE Lista as marcas de veículos referente ao tipo de veículo:
url = "https://brasilapi.com.br/api/fipe/marcas/v1/"
response = requests.get(url)
```

```python
# Verificar acesso a API1
if response.status_code == 200:
    data_json = response.json()
else:
    print("Erro ao acessar a API")
```

```python
# Extrair dados API1
data_json
```

```python
# Extrair dados API1 em tabela
tipos_full = pd.DataFrame(data_json)
tipos_full.head(2)
```

```python
# API2 Retorna informações de todos os bancos do Brasil.
url = "https://brasilapi.com.br/api/banks/v1"
response = requests.get(url)
```

```python
# Verificar acesso a API2
if response.status_code == 200:
    data_json = response.json()
else:
    print("Erro ao acessar a API")
```

```python
# Dados API2
data_json
```

```python
# Tabela API2
banco_full = pd.DataFrame(data_json)
banco_full.head(2)
```
